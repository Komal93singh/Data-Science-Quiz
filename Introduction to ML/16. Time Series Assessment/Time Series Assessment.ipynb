### Q1.  What is the main purpose of the Augmented Dickey-Fuller (ADF) test? 
1. To check for stationery in a time series 
2. To test for autocorrelation 
3. To identify seasonality 
4. To measure trend strength

The main purpose of the Augmented Dickey-Fuller (ADF) test is:

**1. To check for stationarity in a time series**

The ADF test helps determine whether a time series has a unit root, which implies non-stationarity. If the test rejects the null hypothesis (presence of a unit root), the series is considered stationary.

### Q13. Which of the following is NOT a common method for making a time series stationary?
