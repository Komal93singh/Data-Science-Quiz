### Q1. What is TensorFlow primarily used for?

2. Data Analysis

The correct answer is **1. Data Analysis** — but more precisely, TensorFlow is primarily used for **machine learning and deep learning** tasks, such as training neural networks for image recognition, NLP, and predictive modeling.

### Q2. Which library's primary focus is on deep learning and neural network training? 
1. TensorFlow 
2. PyTorch 
3. Scikit-learn 
4. XgBoost

The correct answer is **1. TensorFlow** and **2. PyTorch** — both are primarily focused on deep learning and neural network training.

If you have to pick **only one** in a single-answer question, **PyTorch** is often highlighted for deep learning research, while **TensorFlow** is widely used for both research and production.

For an exam-type question, **TensorFlow** (option 1) would usually be the intended answer.


### Q3. Which library offers strong support for both research and production environments? 
1. TensorFlow 
2. PyTorch 
3. Scikit-learn 
4. Pandas

The correct answer is **1. TensorFlow**

TensorFlow is designed with strong support for **both research and production** — it has tools like TensorFlow Serving, TensorFlow Lite, and TensorFlow.js for deployment across servers, mobile, and browsers, while still being flexible enough for research.

PyTorch is excellent for research but only recently started improving its production capabilities.


### Q4. Which library provides TensorBoard for visualizing the training process? 
1. TensorFlow 
2. PyTorch 
3. Keras 
4. Scikit-learn

The correct answer is **1. TensorFlow** 

**TensorBoard** is TensorFlow’s built-in visualization tool for tracking metrics like loss, accuracy, histograms, and model graphs during training.


### Q5. Which API is not a part of PyTorch? 
1. torch.nn 
2. torch.optim 
3. tf.data 
4. torch.autograd

The correct answer is **3. tf.data**

`tf.data` is part of **TensorFlow**, not PyTorch.

PyTorch has its own equivalents, like `torch.utils.data.Dataset` and `torch.utils.data.DataLoader`.


### Q6. Which of the following allows defining custom gradients in neural networks? 
1. TensorFlow 
2. PyTorch 
3. Both TensorFlow and PyTorch 
4. Neither

The correct answer is **3. Both TensorFlow and PyTorch** 

Both frameworks allow you to define **custom gradients**:

* **TensorFlow** → using `tf.custom_gradient`
* **PyTorch** → by creating custom `autograd.Function` classes and defining `forward` & `backward` methods.

### Q8. What is the purpose of DataLoader in PyTorch?

### Q11. What is a Tensor in TensorFlow? 
1. A special type of neural network 
2. A multi-dimensional array 
3. A loss function

The correct answer is **2. A multi-dimensional array**

In TensorFlow, a **Tensor** is the basic data structure used to represent data — it’s essentially a **multi-dimensional array** (similar to NumPy arrays) that can run on CPUs, GPUs, or TPUs.


### Q12.. How do you install TensorFlow in a Python environment?

The correct answer is **4. All of the above**

### Q19. Which function is used in PyTorch to create a tensor? 
1. torch. Tensor() 
2. np.array() 
3. pd.DataFrame()

### Q20. In PyTorch, what is the purpose of the nn.Module class? 
1. To define a custom loss function 
2. To define a neural network model 
3. To optimize a neural network 
4. To load and preprocess data

The correct answer is **2. To define a neural network model**

In PyTorch, **`nn.Module`** is the base class for **all neural network models**.
You define your layers in `__init__()` and the forward pass logic in `forward()`.


### Q21. Which optimizer is commonly used in PyTorch for training neural networks? 
1. Adam 
2. SGD 
3. RMSprop 
4. All of the above

The correct answer is **4. All of the above**

In PyTorch, you can use multiple optimizers depending on your needs:

* **Adam** → adaptive learning rates, good default choice
* **SGD** → stochastic gradient descent, often with momentum
* **RMSprop** → good for recurrent neural networks and noisy data


### Q22. What is the use of torch.autograd in PyTorch? 
1. Data loading 
2. Automatic differentiation 
3. Model evaluation 
4. Visualizing graphs

The correct answer is **2. Automatic differentiation** 

In PyTorch, **`torch.autograd`** automatically computes **gradients** for tensors during backpropagation, which is essential for training neural networks.


### Q23. Which function is PyTorch is used for backpropagation? 
1. backward() 
2. step() 
3. zero_grad() 
4. forward()

The correct answer is **1. `backward()`**

In PyTorch, calling **`loss.backward()`** computes the gradients of the loss with respect to the model parameters for backpropagation.


### Q24. Which module in PyTorch is used to load an dpreprocess datasets? 
1. torch.utils.data

The correct answer is **1. `torch.utils.data`**

### Q26. Which library provides eager execution by default? 
1. TensorFlow 
2. PyTorch 
3. Both TensorFlow and PyTorch 
4. Neither

The correct answer is **2. PyTorch**

PyTorch runs in **eager execution mode by default**, meaning operations execute immediately as they’re called, which is more intuitive for debugging.

TensorFlow only enabled eager execution by default starting from **TF 2.x** — before that, it used static computational graphs.


### Q27. Which library is known for dynamic computational graphs? 
1. TensorFlow 
2. PyTorch 
3. Keras 
4. Scikit-learn

The correct answer is **2. PyTorch**

PyTorch is famous for its **dynamic computational graphs** (also called "define-by-run"), meaning the graph is built **on the fly** during execution, making it more flexible and easier to debug.

1. TensorFlow

3. Scikit-learn 
4. Matplotlib

The correct answer is **1. TensorFlow**

**TensorFlow** is often preferred for **large-scale, production-level deployment** because it has:

* **TensorFlow.js** for web deployment

### Q29. Which library uses the Sequential model API for simple stack-based networks? 
1. TensorFlow 
2. PyTorch 
3. Both TensorFlow and PyTorch
