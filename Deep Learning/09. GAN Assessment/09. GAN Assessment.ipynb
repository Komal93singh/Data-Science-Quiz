### Q1. Which of the following is NOT a typical application of GANs? 
1. Image generation 
2. Style transfer 
3. Lossless data compression 
4. Data augmentation

The correct answer is:

**3. Lossless data compression**

GANs (Generative Adversarial Networks) are typically used for tasks like image generation, style transfer, and data augmentation, but they are not generally used for lossless data compression.


### Q2. Which of the following statements about the constant input in StyleGANs is correct? 
1. It's a random noise vector provided at the start of generation 
2. It's a learned constant input that the generator starts from 
3. It's the output of the mapping network 
4. It's the final generated image

The correct answer is:

**2. It's a learned constant input that the generator starts from**

In StyleGANs, the generator begins with a learned constant tensor (constant input), which is then progressively transformed by the network layers. This constant input is not random noise but a fixed learned parameter.


### Q3. Which of the following best describes the concept of "latent space" in GANs? 
1. They physical space where the GAN is trained 
2. The set of all possible input images 
3. The high-dimensional space where input vectors for the generator are sampled from  
4. The output space of the discriminator

The correct answer is:

**3. The high-dimensional space where input vectors for the generator are sampled from**

In GANs, the latent space is a high-dimensional vector space from which the generator samples input vectors (usually random noise) to produce generated data like images.


### Q4. Which of the following is NOT a typical application of StyleGANs? 
1. High-quality image synthesis 
2. Style transfer 
3. Lossless data compression 
4. Controllable image generation

The correct answer is:

StyleGANs are widely used for high-quality image synthesis, style transfer, and controllable image generation, but they are not typically applied to lossless data compression.


### Q5. What is the main advantage of using a "disentangled representation" in GANs? 
1. It allows for faster training 
2. It enables independent control over different aspect of the generated output 
3. It eliminates the need for a discriminator 
4. It guarantees perfect reconstruction of input data

The correct answer is:

**2. It enables independent control over different aspects of the generated output**

Disentangled representations allow the model to control separate factors of variation (like pose, color, shape) independently in the generated data, making manipulation and interpretation easier.


### Q6. What is the purpose of the "noise inputs" in StyleGANs? 
1. To destabilize the training process 
2. To generate stochastic details separately from high-level features 
3. To compress the input data 
4. To classify the generated images

The correct answer is:

**2. To generate stochastic details separately from high-level features**

In StyleGANs, noise inputs add fine-grained, random details (like texture variations) to the generated images, allowing the model to separate stochastic variations from overall high-level style and structure.
In StyleGANs, explicit noise inputs are incorporated at each layer of the synthesis network. This allows the model to generate stochastic details (like hair placement, skin pores, etc.) separately from high-level features controlled by the styles.


### Q7. WHich of the following is a key difference between DCGANs and traditional GANs? 
1. DCGANs don't use a generator network 
2. DCGANs use fully connected layers instead of convolutional layers 
3. DCGANs incorporate convolutional layers in both generator and discriminator 
4. DCGANs don't require training data

The correct answer is:

**3. DCGANs incorporate convolutional layers in both generator and discriminator**

DCGANs (Deep Convolutional GANs) use convolutional layers to improve the quality and stability of generated images, unlike traditional GANs that often relied on fully connected layers.


### Q8. What is the purpose of the AdaIN (Adaptive Instance Normalization) operations in StyleGANs? 
1. To classify input images 
2. To generate random noise 
3. To apply style information at different resolutions 
4. To compress the latent space

The correct answer is:

**3. To apply style information at different resolutions**

AdaIN in StyleGANs modulates the feature maps by adjusting their mean and variance according to the style vector, allowing the model to inject style information at various layers and resolutions in the generator.
In StyleGANs, AdaIN operations are used after each convolution layer to apply style information derived from the intermediate latent code (w) to control the generator's behavior at different resolutions.


### Q9. WHich of the following best describes the "progressive growing" technique used in some GAN architectures? 
1. Gradually increasing the number of layers in the generator 
2. Starting from low-resolution images and gradually increasing the resolution during training 
3. Progressively increasing the batch size during training 
4. Gradually decreasing the learning rate over time

The correct answer is:

**2. Starting from low-resolution images and gradually increasing the resolution during training**

Progressive growing trains GANs by first generating low-resolution images and then incrementally adding layers to both generator and discriminator to handle higher resolutions, which stabilizes training and improves image quality.


### Q10. What is the primary purpose of the "critic" in a Wasserstein GAN? 
1. To generate new data samples 
2. To estimate the Wasserstein distance between real and generated distributions 
3. To compress the input data 
4. To perform style transfer

The correct answer is:

**2. To estimate the Wasserstein distance between real and generated distributions**

In a Wasserstein GAN, the "critic" replaces the discriminator and provides a continuous score that helps measure how close the generated data distribution is to the real data distribution, using the Wasserstein (Earth Mover’s) distance.


### Q11. WHich of the following is NOT a technique for achieving controllable generation in GANs? 
1. Conditional GANs (cGANs) 
2. InfoGAN 
3. Gradient clipping 
4. StyleGAN

The correct answer is:

**3. Gradient clipping**

Gradient clipping is a technique used to stabilize training by preventing exploding gradients, but it is not a method for achieving controllable generation in GANs. The other options (Conditional GANs, InfoGAN, and StyleGAN) are designed to enable control over generated outputs.

Controllable generation means the ability to steer the GAN to generate outputs with desired attributes or features, such as specific styles, classes, or characteristics.

### Q15. What is the purpose of the mapping network in StyleGANs? 
1. To classify input images 
2. To transform the input latent code into an intermediate latent space 
3. To generate the final output image 
4. To normalize the input data

The correct answer is:

**2. To transform the input latent code into an intermediate latent space**

In StyleGANs, the mapping network takes the initial latent vector ( z ) and maps it to an intermediate latent space ( w ), which is then used to control styles in the synthesis network more effectively and disentangle factors of variation.


### Q16. Which of the following is a key feature of Deep Convolutional GANs (DCGANs)? 
1. Using fully connected layers throughout the network 
2. Avoiding batch normalization 
3. Using ReLU activation in the discriminator 
4. Incorporating convolutional and transposed convolutional layers

The correct answer is:

**4. Incorporating convolutional and transposed convolutional layers**

DCGANs use convolutional layers in the discriminator and transposed convolutional (also called deconvolutional) layers in the generator to effectively model image data.


### Q17. In the context of GANs, what does "mode collapse" refer to? 
1. When the discriminator always classifies inputs as fake 
2. When the generator produces a limited variety of samples 
3. When the training process becomes unstable 
4. When the GAN fails to converge

The correct answer is:

**2. When the generator produces a limited variety of samples**

Mode collapse happens when the generator gets stuck producing only a few similar outputs instead of diverse samples, limiting the GAN’s ability to represent the full data distribution.

### Q22. Which of the following best describes the "minimax game" in the context of GANs? 
1. A game where the generator tries to minimize its loss while the discrimintor tries to maximize it 
2. A technique to minimize the size of the neural network 
3. A method to maximize the training speed 
4. A game where both newtroks try to minimize their respective losses

The correct answer is:

**1. A game where the generator tries to minimize its loss while the discriminator tries to maximize it**

In GANs, the generator and discriminator compete in a minimax game where the generator aims to fool the discriminator, minimizing its loss, while the discriminator tries to correctly distinguish real from fake, maximizing its ability.


### Q23. In the context of WGANs, what is the purpose of the "gradient penalty"? 
1. To speed up the training process 
2. To enforce Lipschitz continuity on the critic 
3. To generate higher quality images 
4. To eliminate the need for batch normalization

The correct answer is:

**2. To enforce Lipschitz continuity on the critic**

The gradient penalty helps maintain the Lipschitz constraint in WGANs, which is essential for stable training and ensuring the Wasserstein distance is properly estimated.

The correct answer is:

**3. To encourage localization of style effects and better control**

Style mixing in StyleGANs combines styles from different latent codes at various layers during training to promote disentanglement and localized control over different image features.
