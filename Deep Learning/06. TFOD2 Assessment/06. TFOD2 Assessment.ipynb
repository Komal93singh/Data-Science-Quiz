### Q1. What is TensorFlow object Detection API (TFOD2) designed for? 
1. Image classification 
2. Object detection 
3. Natural language processing 
4. Data annotation

The correct answer is:

**2. Object detection**

TensorFlow Object Detection API (TFOD2) is specifically designed to help build, train, and deploy object detection models.

**1. The number of object classes in the dataset**

### Q10. In TFOD2, what does the Mean Average Precision (mAP) metric evaluate?
