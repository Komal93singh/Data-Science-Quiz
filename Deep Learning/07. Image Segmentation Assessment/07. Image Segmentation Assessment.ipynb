### Q1. Which evaluation metric measures the overlap between predicted and ground truth segmentation masks? 
1. Accuracy 
2. Precision 
3. Recall 
4. Intersection over Union (IoU)

The evaluation metric that measures the overlap between predicted and ground truth segmentation masks is:

**4. Intersection over Union (IoU)**

IoU calculates the ratio of the area of overlap between the predicted mask and the ground truth mask to the area of their union, making it a standard metric for segmentation tasks.


### Q2. Which of the following is a type of image segmentation? 
1. Color segmentation 
2. Instance segmentation 
3. Edge detection 
4. Object detection

The correct answer is:

**2. Instance segmentation**

Instance segmentation is a type of image segmentation where each individual object instance is segmented separately.

* Color segmentation is more of a basic technique or preprocessing step.
* Edge detection is a feature extraction technique, not segmentation itself.
* Object detection involves identifying objects with bounding boxes, not pixel-wise segmentation.


### Q3. Which of the following is NOT a common data augmentation technique for image segmentation? 
1. Random cropping 
2. Horizontal flipping 
3. Color jittering 
4. Gaussian blurring

The answer is:

**4. Gaussian blurring**

Gaussian blurring is generally **not** a common data augmentation technique for image segmentation because it can degrade important details needed for accurate segmentation. The others—random cropping, horizontal flipping, and color jittering—are commonly used to increase data diversity without significantly harming the key features.


### Q4. Which of the following is a charcteristics of instance segmentation? 
1. It only works on binary images 
2. It assigns a unique identifier to each object instance 
3. It cannot handle overlapping objects 
4. It requires manual annotation for every pixel

The correct answer is:

**2. It assigns a unique identifier to each object instance**

Instance segmentation distinguishes between different instances of the same class by assigning unique IDs, allowing it to handle overlapping objects as well.


### Q5. Which of the following is NOT a common post-processing technique for refining segmentation masks? 
1. Morphological operations 
2. Conditional random fields 
3. Fourier transform 
4. Connected component analysis

The answer is:

**3. Fourier transform**

Fourier transform is not commonly used as a post-processing technique for refining segmentation masks. The other options—morphological operations, conditional random fields, and connected component analysis—are widely used to improve segmentation results.


### Q6. Which of the following is a characteristic of semantic segmentation? 
1. It assigns unique labels to each object instance 
2. It only works on grayscale images 
3. It assigns the same label to all pixels of the same class 
4. It requires 3D image data

The correct answer is:

**3. It assigns the same label to all pixels of the same class**

Semantic segmentation labels every pixel according to the class it belongs to, without distinguishing between different object instances of the same class.


### Q7. What is the main purpose of transform learning in image segmentation? 
1. To reduce model size 
2. To improve training speed 
3. To leverage pre-trained features 
4. To eliminate teh need for data annotation

The correct answer is:

**3. To leverage pre-trained features**

Transfer learning uses features learned from a pre-trained model (often on large datasets) to improve performance and reduce the amount of training data needed for a new segmentation task.


### Q8. Which of the following is NOT a common technique for handling class imbalance in segmentation datasets? 
1. Oversampling 
2. Undersampling 
3. Weighted loss function 
4. Gradient clipping

The answer is:

**4. Gradient clipping**

Gradient clipping is used to prevent exploding gradients during training, not specifically to handle class imbalance. Oversampling, undersampling, and weighted loss functions are common techniques for dealing with class imbalance in segmentation datasets.


### Q9. What is the primary function of the backbone network in Mask R-CNN? 
1. Object classification 
2. Bounding box regression 
3. Feature extraction 
4. Mask generation

The correct answer is:

**3. Feature extraction**

The backbone network in Mask R-CNN (often ResNet or a similar CNN) is responsible for extracting rich feature representations from the input image, which are then used for classification, bounding box regression, and mask generation.


### Q10. Which of the following is a key advantage of Mask R-CNN over Faster R-CNN? 
1. Faster inference speed 
2. Lower memory requirements 
3. Ability to perform instance segmentation 
4. Better object classification accuracy

The correct answer is:

**3. Ability to perform instance segmentation**

Mask R-CNN extends Faster R-CNN by adding a branch for predicting segmentation masks on each detected object, enabling instance segmentation, which Faster R-CNN does not do.


### Q11. What is the purpose of non-maximum suppression in object detection? 
1. To increase the number of detected objects 
2. To eliminate redundant detections 
3. To improve object classification accuracy 
4. To generate segmentation masks

The correct answer is:

**2. To eliminate redundant detections**

Non-maximum suppression (NMS) removes overlapping bounding boxes that correspond to the same object, keeping only the one with the highest confidence score.


### Q12. Which of the following is NOT a common preprocessing step for image segmentation? 
1. Resizing 
2. Normalization 
3. Data encryption 
4. Color space conversion

The answer is:

**3. Data encryption**

Data encryption is not a preprocessing step for image segmentation. Resizing, normalization, and color space conversion are commonly used preprocessing steps.


### Q13. What is the primary goal of panoptic segmentation? 
1. To classify pixels into semantic categories 
2. To detect individual object instances 
3. To combine semantic and instance segmentation 
4. To perform 3D reconstruction

The correct answer is:

**3. To combine semantic and instance segmentation**

Panoptic segmentation unifies semantic segmentation (classifying every pixel) and instance segmentation (differentiating individual object instances) into a single task.


### Q14. Which of the following is NOT a type of image segmentation? 
1. Semantic segmentation 
2. Instance segmentation 
3. Panoptic segmentation 
4. Volumetric segmentation

The answer is:

**4. Volumetric segmentation**

Volumetric segmentation typically refers to segmentation in 3D data (like medical imaging volumes), not a standard type of 2D image segmentation. The other three are recognized types of image segmentation.


### Q15. What is the purpose of data augmentation in training segmentation models? 
1. To reduce the size of the dataset 
2. To increase model complexity 
3. To improve model generalization 
4. To speed up training

The correct answer is:

**3. To improve model generalization**

Data augmentation artificially increases the diversity of the training dataset, helping the model generalize better to unseen data.


### Q16. Which of the following is a common challenge in image segmentation? 
1. Occlusions 
2. Varying lighting conditions 
3. Complex backgrounds 
4. All of the above

The correct answer is:

**4. All of the above**

Occlusions, varying lighting conditions, and complex backgrounds are all common challenges that make image segmentation difficult.


### Q17. What is the main advantage of using polygon masks over bounding boxes? 
1. Faster processing 
2. Lower memory requirements 
3. More accurate object representation 
4. Easier to implement

The correct answer is:

**3. More accurate object representation**

Polygon masks precisely outline the shape of an object, providing a more accurate segmentation compared to bounding boxes, which only enclose the object in a rectangle.


### Q18. Which of the following is NOT a common application of image segmentation? 
1. Medical imaging 
2. Autonomous driving 
3. Speech recognition 
4. Satellite imagery analysis

The answer is:

Speech recognition deals with audio data, not images, so it doesn’t use image segmentation. The other options are common applications of image segmentation.


### Q19. What is the purpose of RoI Align in Mask R-CNN? 
1. To classify objects 
2. To generate bounding boxes 
3. To preserve spatial information 
4. To create binary masks

The correct answer is:

**3. To preserve spatial information**

RoI Align improves on RoI Pooling by avoiding quantization, which helps preserve precise spatial locations, leading to better mask predictions in Mask R-CNN.


### Q20. Which component of Mask R-CNN is responsible for generating object proposals? 
1. Backbone network 
2. RoI Align 
3. Region Proposal Network (RPN) 
4. MAsk head

The correct answer is:

**3. Region Proposal Network (RPN)**

The RPN generates object proposals (candidate bounding boxes) that are then processed by the rest of the Mask R-CNN pipeline.


### Q21. WHat is the primary difference between semantic and instance segmentation? 
1. Semantic segmentation is faster 
2. Instance segmentation distinguishes between objects of the same class 
3. Semantic segmentation uses deep learning, while instance segmentation doesn't 
4. Instance segmentation only works on color images

The correct answer is:

**2. Instance segmentation distinguishes between objects of the same class**

Semantic segmentation labels all pixels of the same class identically, while instance segmentation differentiates between individual object instances within the same class.


### Q22. WHat is the primary advantage of using deep learning models for image segmentation? 
1. They require less training data 4
2. They are always faster the traditional methods 
3. They can learn complex features automatically 
4. They don't require any preprocessing

The correct answer is:

**3. They can learn complex features automatically**

Deep learning models can automatically learn hierarchical and complex features from data, improving segmentation accuracy without manual feature engineering.


### Q23. What is teh purpose of anchor boxes in object detection and segmentation? 
1. To improve image resolution 
2. To provide reference sizes and aspect ratios for object proposals 
3. To perform data augmentation 
4. To calculate loss functions

The correct answer is:

**2. To provide reference sizes and aspect ratios for object proposals**

Anchor boxes are predefined boxes of various sizes and aspect ratios that help the model propose bounding boxes more effectively during detection and segmentation.


### Q24. WHat is the main advantage of using a Feature Pyramid Network (FPN) in segamentation models? 
1. It reduces the number of model parameters 
2. It improves teh detection of objects at different scales 
3. It eliminates the need for data augmentation 
4. It guarantees real-time performance

The correct answer is:

**2. It improves the detection of objects at different scales**

Feature Pyramid Networks (FPNs) combine features from multiple layers to help the model detect objects of varying sizes more effectively.


### Q25. What is the primary purpose of the mask head in Mask R-CNN? 
1. To classify objects 
2. To generate bounding boxes 
3. To produce segmentation masks 
4. To perform feature extraction

The correct answer is:

**3. To produce segmentation masks**

The mask head in Mask R-CNN is responsible for predicting the pixel-wise segmentation mask for each detected object.
