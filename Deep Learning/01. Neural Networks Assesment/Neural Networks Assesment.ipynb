### Q29. What is the purpose of zero initialization in neural networks? 
1. To prevent symmetry in the model 
2. To ensure identical weight updates during training 
3. To simplify model training 
4. To initialize biases

The purpose of zero initialization in neural networks is:

**2. To ensure identical weight updates during training**

However, this is actually a problem — zero initialization of weights causes all neurons to learn the same features (symmetry problem). So usually, zero initialization is avoided for weights but can be used for biases.

If you want the best option for zero initialization's known effect, it’s:

**2. To ensure identical weight updates during training** (which is why it’s generally avoided for weights).
