### Q7. What does the following regex pattern match: '^\d{3}-\d{2}-\d{4}$'? 
1. A date in the format YYYY-MM-DD
2. A phone number 
3. A social security number 
4. A hexadecimal color code

The correct answer is:

**3. A social security number** 

Because:

* `^` → start of the string
* `\d{3}` → exactly 3 digits
* `-` → a hyphen
* `\d{2}` → exactly 2 digits
* `-` → a hyphen
* `\d{4}` → exactly 4 digits
* `$` → end of the string

This matches the format **XXX-XX-XXXX**, which is the standard U.S. Social Security Number format.
